# 04 - Evaluación

Notebook reservado para la Fase 4 y posteriores: reconstrucción y comparación de resultados.

```python
import sys
from pathlib import Path
import numpy as np

sys.path.append("..")
from src.audio_loader import load_audio
from src.metrics import mean_squared_error, signal_to_noise_ratio

# 2. Cargamos los audios que guardamos en la fase de filtrado
RUTA_RUIDO = Path("../data/processed/1_audio_con_zumbido.wav")
RUTA_FILTRADO = Path("../data/processed/2_audio_limpio_fft.wav")

senal_ruidosa, sample_rate = load_audio(RUTA_RUIDO)
senal_filtrada, _ = load_audio(RUTA_FILTRADO)

# 3. Nuestra referencia ideal: el tono de 440 Hz
# Generamos matemáticamente el tono puro de 440 Hz exacto que teníamos al inicio
t = np.arange(len(senal_ruidosa)) / sample_rate
senal_ideal = 0.5 * np.sin(2 * np.pi * 440 * t)

# 4. APLICAMOS LAS MÉTRICAS DE TU MÓDULO (Fase 5)
# Evaluación ANTES del filtro
snr_antes = signal_to_noise_ratio(senal_ideal, senal_ruidosa)
mse_antes = mean_squared_error(senal_ideal, senal_ruidosa)

# Evaluación DESPUÉS del filtro
snr_despues = signal_to_noise_ratio(senal_ideal, senal_filtrada)
mse_despues = mean_squared_error(senal_ideal, senal_filtrada)

# 5. IMPRESIÓN DE RESULTADOS
print("=== REPORTE FINAL DE MÉTRICAS (FASE 5) ===")
print("\n[ RELACIÓN SEÑAL-RUIDO (SNR) ] ")
print(f"  Antes del filtro:   {snr_antes:7.2f} dB")
print(f"  Después del filtro: {snr_despues:7.2f} dB")

print("\n[ ERROR CUADRÁTICO MEDIO (MSE) ] ")
print(f"  Antes del filtro:   {mse_antes:.6f}")
print(f"  Después del filtro: {mse_despues:.6f}")
```

## Evaluación con métricas numéricas

Escuchar el audio nos dice si algo mejoró, pero no nos dice *cuánto* mejoró. Para eso usamos métricas matemáticas que cuantifican la diferencia entre la señal ideal y la señal procesada.

Usamos dos métricas:

### MSE — Error Cuadrático Medio
Mide qué tan diferente es la señal filtrada de la señal ideal. **Más cercano a 0 es mejor.** Un MSE de 0 significaría que la señal filtrada es perfectamente igual a la original limpia.

### SNR — Relación Señal-Ruido (en dB)
Mide qué tan fuerte es la señal útil comparada con el ruido residual. **Más alto es mejor.** Se expresa en decibeles (dB):
- Valores negativos → el ruido es más fuerte que la señal
- Entre 0 y 10 dB → señal débil con bastante ruido
- Por encima de 20 dB → señal notablemente más fuerte que el ruido

Comparamos ambas métricas **antes y después del filtro** para demostrar numéricamente que el procesamiento mejoró la señal.
