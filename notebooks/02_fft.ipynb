Notebook reservado para la Fase 2: cálculo de FFT y análisis espectral.

```python
import sys
from pathlib import Path
import matplotlib.pyplot as plt

# Añadimos la carpeta raíz al path para usar nuestra arquitectura modular
sys.path.append("..")
from src.audio_loader import load_audio
from src.fft_utils import calcular_fft

# 1. Cargamos el audio sintético que generamos anteriormente
AUDIO_PATH = Path("../data/raw/PruebaAN.wav")
signal, sample_rate = load_audio(AUDIO_PATH)

# 2. Aplicamos la FFT utilizando nuestro nuevo módulo
frecuencias, magnitud = calcular_fft(signal, sample_rate)

# 3. Visualizamos el Espectro de Frecuencias
plt.figure(figsize=(12, 4))
plt.plot(frecuencias, magnitud, color='darkorange', linewidth=1)
plt.title("Espectro de Frecuencias (Dominio Frecuencial)")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud")

# Hacemos "zoom" limitando el eje X de 0 a 1000 Hz, 
# ya que nuestras frecuencias de interés están ahí.
plt.xlim(0, 1000) 
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Paso 1: Aplicar la FFT y ver el espectro de frecuencias

Aquí ocurre la transformación más importante del proyecto: convertir la señal del **dominio temporal** al **dominio frecuencial**.

Imagina que tienes un batido de frutas. En el dominio temporal solo ves el líquido mezclado. La FFT actúa como si desarmara ese batido y te dijera exactamente cuánta fresa, cuánto mango y cuánta piña hay dentro.

En términos de audio:
- El eje **X** (horizontal) muestra las **frecuencias** presentes en la señal, medidas en Hz.
- El eje **Y** (vertical) muestra la **magnitud** (qué tan intensa es cada frecuencia).

Lo que esperamos ver en la gráfica son dos picos muy claros:
- Uno en **60 Hz** → el zumbido eléctrico que inyectamos
- Uno en **440 Hz** → la nota La que creamos

Si los ves, significa que la FFT está funcionando correctamente.
